LSTM PREDICTION MODEL (72 HOURS PAST DATA - 24 HOURS FUTURE PREDICTION)

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, accuracy_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
#Select Features
features = [
    'PM2.5', 'PM10', 'TEMPERATURE', 'HUMIDITY',
    'WS', 'WD', 'RF', 'Ozone'
]

data = merged_df[features]
```

```python
#Normalize Data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
```

```python
#Create Sequence
def create_sequences(data, window=72, horizon=24):
    X, y = [], []
    
    for i in range(len(data) - window - horizon):
        X.append(data[i:i+window])
        y.append(data[i+window:i+window+horizon, 0:2])  # PM2.5, PM10
    
    return np.array(X), np.array(y)

X, y = create_sequences(scaled_data)
```

```python
#Train-Test Split
split = int(len(X) * 0.8)

X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
#Build LSTM Model
from tensorflow.keras.layers import Input

model = Sequential([
    Input(shape=(72, X.shape[2])),

    LSTM(128, return_sequences=True),
    Dropout(0.2),

    LSTM(64),
    Dropout(0.2),

    Dense(32, activation='relu'),
    Dense(48)
])
```

```python
#Compile Model
model.compile(
    optimizer='adam',
    loss='mse'
)
```

```python
#Train Model
model.fit(
    X_train,
    y_train.reshape(y_train.shape[0], -1),
    epochs=20,
    batch_size=32,
    validation_data=(X_test, y_test.reshape(y_test.shape[0], -1))
)
```

```python
#Predict
pred = model.predict(X_test)

# reshape back to (samples, 24, 2)
pred = pred.reshape(-1, 24, 2)
y_test = y_test.reshape(-1, 24, 2)
```

```python
#Evaluate
mae_pm25 = mean_absolute_error(y_test[:,:,0].flatten(), pred[:,:,0].flatten())
mae_pm10 = mean_absolute_error(y_test[:,:,1].flatten(), pred[:,:,1].flatten())

print("PM2.5 MAE:", mae_pm25)
print("PM10 MAE:", mae_pm10)
```

```python
#Convert to Work/Not
PM25_LIMIT = 100
PM10_LIMIT = 250

def decision(pm25, pm10):
    return 0 if (pm25 > PM25_LIMIT or pm10 > PM10_LIMIT) else 1
```

```python
#Decision Accuracy
actual = []
predicted = []

for i in range(len(pred)):
    for t in range(24):
        actual.append(decision(y_test[i, t, 0], y_test[i, t, 1]))
        predicted.append(decision(pred[i, t, 0], pred[i, t, 1]))

from sklearn.metrics import accuracy_score, classification_report

print("Decision Accuracy:", accuracy_score(actual, predicted))
print(classification_report(actual, predicted))
```

```python
#Real-Time Prediction
# Take last 72 hours
last_72 = scaled_data[-72:]

# reshape for model
last_72 = last_72.reshape(1, 72, X.shape[2])

future = model.predict(last_72)

future = future.reshape(24, 2)

# Example: next day average PM
pm25_next = future[:,0].mean()
pm10_next = future[:,1].mean()

print("Tomorrow Avg PM2.5:", pm25_next)
print("Tomorrow Avg PM10:", pm10_next)

if decision(pm25_next, pm10_next):
    print("✅ CONTINUE WORK")
else:
    print("❌ STOP WORK")
```
